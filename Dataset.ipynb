```python
path = kagglehub.dataset_download("camnugent/california-housing-prices")

print("Path:", path)

# Build full file path
file_path = os.path.join(path, "housing.csv")

# Load into DataFrame
df = pd.read_csv(file_path)

df.isnull().sum(), df.notnull()
```

```python
import matplotlib.pyplot as plt

df.hist(figsize=(12,8), bins=30)# x axis is value range
plt.tight_layout()                # y axis is frequency
plt.show()
```

```python
df["median_income"].hist(bins=30)
plt.xlabel("Median Income") #average income of the family in that area basically in $10,000
plt.ylabel("Frequency")     #no. of areas in that income
plt.show()
```

```python
plt.scatter(df["longitude"], df["latitude"],
            c=df["median_house_value"],
            cmap="jet",
            alpha=0.4)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label="House Value")
plt.show()
```
