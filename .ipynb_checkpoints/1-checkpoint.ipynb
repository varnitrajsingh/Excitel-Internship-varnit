```python
class point:
    __match_args__ = ('x','y')
    def __init__(self,x,y):
        self.x = x
        self.y = y
match point:
    case[]:
        print("no point")
    case[point(0,0)]:
        print("point is at the origin")
    case[point(x,y)]:
        point(f"the point is at {x} , {y}")
    case[point(0,y1) , point(0,y2)]:
        print(f"the point is at {y1}, {y2}")
    case _:
        print("something else")
    
```

```python
day = "Friday"

match day:
    case "Monday":
        print('Start of the week')
    case "Wednesday":
        print('Mid of the week')
    case _:
        print('default case')
```

```python
point = (5,9)

match point:
    case (0,0):
        print('the origin')
    case (x,0):
        print(f"the point is on X-axis at {x}")
    case (0,y):
        print(f"the point is on Y-axis at {y}")
    case (x,y):
        print(f"the point is {x} times on x-axis and {y} times on y-axis")
    case _:
        print("Nothing")
```

```python
from enum import Enum
class Traffic(Enum):
    red = "red"
    yellow = "yellow"
    green = "green"

choice = input('Enter the requirement').lower()
demand = Traffic(choice)

if demand == Traffic.red:
    print('stop')
elif demand == Traffic.green:
    print('go')
elif demand == Traffic.yellow:
    print('wait')
```

```python
def fib(n):
    result = []
    a,b = 0,1
    while a<n:
        result.append(a)
        a,b = b,a+b
    return result

fib(100)
```

```python
def fact(n):
    result = 1
    for i in range(1,n+1):
        result *= i
    return result
fact(5)
```

```python
def ask(prompt, retries = 3, reminder='try again later'):
    while True:
        reply = input(prompt).lower()
        
        if reply in {'yes', 'ye', 'y'}:
            return True
        elif reply in {'no', 'n', 'nop', 'nope'}:
            return False
            
        retries -= 1
        if retries < 0:
            raise ValueError("invalid response")
            
        print(reminder)
```

```python
def total(*args):
    tot = 0
    for i in args:
        tot += i
    return tot

print(total(1,4,5))
```
